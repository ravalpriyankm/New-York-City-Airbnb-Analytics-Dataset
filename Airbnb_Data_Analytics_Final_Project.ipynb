## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
## Check for missing values in the dataframe and display the count in ascending order.
missing_values = list_clean.isnull().sum().sort_values(ascending=True)
print("Missing Values Count:")
print(missing_values)
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.
# Impute missing values based on data types of columns
for col in list_clean.columns:
    if list_clean[col].dtype == 'object':
        list_clean[col].fillna('Unknown', inplace=True)  # For object columns, fill missing values with 'Unknown'
    else:
        list_clean[col].fillna(list_clean[col].median(), inplace=True)  # For numerical columns, fill missing values with median
```

```python
list_clean
```

```python
# Check for duplicate rows and remove them
before_duplicates = len(list_clean)
list_clean.drop_duplicates(inplace=True)
after_duplicates = len(list_clean)

# Display total number of records before and after removing duplicates
print(f"Total number of records before removing duplicates: {before_duplicates}")
print(f"Total number of records after removing duplicates: {after_duplicates}")
```

## Task 3: Data Transformation (Any Tool)


- Rename the column `availability 365` to `days_booked`
- Convert all column names to lowercase and replace the spaces in the column names with an underscore "_".
- Remove the dollar sign and comma from the columns `price` and `service_fee`. If necessary, convert these two columns to the appropriate data type.

```python
## Rename the column.
list_clean.rename(columns={'availability 365': 'days_booked'}, inplace=True)
```

```python
## Convert all column names to lowercase and replace the spaces with an underscore "_"
list_clean.columns = list_clean.columns.str.lower().str.replace(' ', '_')
```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.

# Removing dollar sign and comma from 'price' and 'service_fee' columns
list_clean['price'] = list_clean['price'].replace({'\$': '', ',': ''}, regex=True)
list_clean['service_fee'] = list_clean['service_fee'].replace({'\$': '', ',': ''}, regex=True)

# Converting 'price' and 'service_fee' columns to numeric, coercing errors to NaN
list_clean['price'] = pd.to_numeric(list_clean['price'], errors='coerce')
list_clean['service_fee'] = pd.to_numeric(list_clean['service_fee'], errors='coerce')
```

## Task 4: Exploratory Data Analysis

- List the count of various room types avaliable in the dataset.
- Which room type has the most strict cancellation policy?
- List the average price per neighborhood group, and highlight the most expensive neighborhood to rent from.

```python
corrections = {
    'brookln': 'Brooklyn',
    'manhatan': 'Manhattan'
}
list_clean['neighbourhood_group'] = list_clean['neighbourhood_group'].replace(corrections)
list_clean['neighbourhood_group'].unique()
```

```python
## List the count of various room types avaliable with Airbnb
room_type_counts = list_clean ['room_type'].value_counts()
print("Count of Various Room Types Available:")
print(room_type_counts)
```

```python
## Which room type adheres to more strict cancellation policy
strict_cancellation = list_clean.groupby('room_type')['cancellation_policy'].value_counts().groupby(level=0).head(1)
print("Room type with the most strict cancellation policy:")
print(strict_cancellation)
```

```python
## List the prices by neighborhood group and also mention which is the most expensive neighborhood group for rentals
avg_price_neighborhood = list_clean.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
print("Average price per neighborhood group:")
print(avg_price_neighborhood)
```

```python
# Highlighting the most expensive neighborhood
most_expensive_neighborhood = avg_price_neighborhood.idxmax()
print("\nThe most expensive neighborhood to rent from is:", most_expensive_neighborhood)
```

```python
# Plotting average price per neighborhood group
plt.figure(figsize=(10, 6))
avg_price_neighborhood.plot(kind='bar', color='skyblue')
plt.title('Average Price per Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='neighbourhood_group', y='price', data=list_clean)
plt.title('Distribution of Prices per Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
list_clean['neighbourhood_group'].value_counts().plot(kind='bar', color='salmon')
plt.title('Count of Listings per Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Count of Listings')
plt.xticks(rotation=45)
plt.show()
```

## Task 5a: Data Visualization

* List the count of various room types avaliable with Airnb
* Which room type adheres to more strict cancellation policy
* List the prices by neighborhood group and also mention which is the most expensive neighborhood group for rentals
* List the top 10 neighborhoods in the increasing order of their price with the help of a horizontal bar graph. Which is the cheapest neighborhood.

```python
# Count of Various Room Types Available with Airbnb
room_type_counts = list_clean['room_type'].value_counts()
print("Count of Various Room Types Available:")
print(room_type_counts)
```

```python
# Room Type with More Strict Cancellation Policy
strict_cancellation = list_clean.groupby('room_type')['cancellation_policy'].value_counts().unstack()
print("Room type with the most strict cancellation policy:")
print(strict_cancellation)
```

```python
# Which room type adheres to a more strict cancellation policy in Bar Graph
strict_cancellation.plot(kind='bar', stacked=True)
plt.title('Cancellation Policy by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.legend(title='Cancellation Policy')
plt.show()
```

```python
# List the prices by neighborhood group and identify the most expensive neighborhood group for rentals
avg_price_neighborhood = list_clean.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
print("Average Price per Neighborhood Group:")
print(avg_price_neighborhood)

most_expensive_neighborhood = avg_price_neighborhood.idxmax()
print(f"The most expensive neighborhood for rentals is: {most_expensive_neighborhood}")
```

```python
# List the top 10 neighborhoods in increasing order of their price with a horizontal bar graph
top_10_neighborhoods = list_clean.groupby('neighbourhood')['price'].mean().sort_values().head(10)
plt.figure(figsize=(10, 6))
top_10_neighborhoods.plot(kind='barh', color='skyblue')
plt.title('Top 10 Neighborhoods by Price (Cheapest to Expensive)')
plt.xlabel('Average Price')
plt.ylabel('Neighborhood')
plt.show()
```

```python
# List the neighborhoods offering short-term rentals within 10 days and illustrate with a bar graph
short_term_rentals = list_clean[list_clean['minimum_nights'] <= 10]['neighbourhood'].value_counts().head(10)
plt.figure(figsize=(10, 6))
short_term_rentals.plot(kind='bar', color='salmon')
plt.title('Neighborhoods Offering Short-Term Rentals (<= 10 days)')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# List the prices with respect to room type using a bar graph and state your inferences
plt.figure(figsize=(8, 6))
sns.barplot(x='room_type', y='price', data=list_clean, ci=None)
plt.title('Prices with Respect to Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

```python
# Create a pie chart showing the distribution of booked days for each neighborhood group
booking_days = list_clean.groupby('neighbourhood_group')['minimum_nights'].sum()
plt.figure(figsize=(8, 8))
pie_chart = booking_days.plot(kind='pie', autopct='%1.1f%%', startangle=140, labeldistance=1.1)
plt.title('Distribution of Booked Days for Each Neighborhood Group')
plt.ylabel('')
plt.show()
```

## Task 5b: Data Visualization

```python
# Filter out missing values in 'review_rate_number' and 'price'
filtered_data_review = list_clean.dropna(subset=['review_rate_number', 'price'])

# Filter out missing values in 'host_identity_verified' and 'price'
filtered_data_host = list_clean.dropna(subset=['host_identity_verified', 'price'])

plt.figure(figsize=(14, 6))

# Violin plot for Review Rate Number vs. Price
plt.subplot(1, 2, 1)
sns.violinplot(x='review_rate_number', y='price', data=filtered_data_review)
plt.title('Effect of Review Rate Number on Price')
plt.xlabel('Review Rate Number')
plt.ylabel('Price')

# Violin plot for Host Identity Verified vs. Price
plt.subplot(1, 2, 2)
sns.violinplot(x='host_identity_verified', y='price', data=filtered_data_host)
plt.title('Effect of Host Identity Verified on Price')
plt.xlabel('Host Identity Verified')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
max_reviews_per_group = list_clean.groupby('neighbourhood_group')['number_of_reviews'].max()
max_reviews_missing = list_clean[list_clean['neighbourhood_group'].isna()]['number_of_reviews'].max()

data = pd.DataFrame({
    'Neighbourhood Group': max_reviews_per_group.index,
    'Max Number of Reviews': max_reviews_per_group.values
})

missing_neighborhood_data = pd.DataFrame({
    'Neighbourhood Group': 'Missing Neighborhoods',
    'Max Number of Reviews': max_reviews_missing
}, index=[0])

data = pd.concat([data, missing_neighborhood_data], ignore_index=True)

display(data)

# Create a bar plot for maximum number of reviews per neighborhood group
plt.figure(figsize=(10, 6))
plt.bar(data['Neighbourhood Group'], data['Max Number of Reviews'], color='skyblue')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Max Number of Reviews')
plt.title('Maximum Number of Reviews per Neighbourhood Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
